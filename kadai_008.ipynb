##**★条件分岐のif文で処理を切り替えよう**

```python
# if文を用いて、以下の条件分岐をするコードを記述してください。値はprint関数で出力してください。

# ・変数varが、3の倍数の場合は「Fizz」を出力
# ・変数varが、5の倍数の場合は「Buzz」を出力
# ・変数varが、3の倍数と5の倍数の両方に該当する場合は「FizzBuzz」を出力
# ・上記のどの場合にも該当しない場合は、変数varの値を出力

# ただし、変数varは正の整数とします。
```

```python
var = 30 #変数定義

if var % 3 == 0 : # n % 3 == 0 で3の倍数の意味
  print("Fizz")
elif var % 5 == 0 : # elif～(あるいは)
  print("Buzz")
elif var % 15 == 0 :
  print("FizzBuzz")
else : #　それ以外は
  print(var)
```
