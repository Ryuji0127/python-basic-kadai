##**課題「繰り返し処理を使いこなそう」**

```python
array = ["水","金","地","火","木","土","天","海","冥"]

for i in array : # for文　arrayを先頭からiに代入する
  print(i) # iの値を出力する

index = 0 # 初期化

while index <= 8: # <= は左辺の値が右辺の値以下の場合True　indexの値が9になるfalse 処理が止まる
  print(array[index])

  index = index + 1 #Loopするごとにindexに +1 される



```
